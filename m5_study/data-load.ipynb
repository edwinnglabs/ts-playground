```python
import pandas as pd
```

```python
calendar_raw = pd.read_csv('./raw/calendar.csv')
response_raw = pd.read_csv('./raw/sales_train_evaluation.csv')
sell_prices_raw = pd.read_csv('./raw/sell_prices.csv')
```

```python
response_df = response_raw.copy()
d_vars = ["d_{}".format(x) for x in range(1, 1942)]
id_vars = response_raw['id'].str.replace('_evaluation', '')
id_vars = list(id_vars)
```

```python
val_mat = response_raw[d_vars].values
val_mat = val_mat.transpose()
val_mat
```

```python
val_mat.shape
```

```python
len(id_vars)
```

```python
response_df = pd.DataFrame(val_mat, columns=id_vars)
response_df['d'] = d_vars
response_df = pd.merge(response_df, calendar_raw[['date', 'd']], on='d')
# response_df = response_df[['date']]
```

```python
response_df[['date'] + id_vars]
```

```python
uni_event1 = calendar_raw['event_name_1'].unique().tolist()
uni_event2 = calendar_raw['event_name_2'].unique().tolist()

# filter nan
uni_event1 = uni_event1[1:]
uni_event2 = uni_event2[2:]
uni_events = list(set(uni_event1 + uni_event2))
```

```python
uni_events
```

```python
event_df = response_df[['date']].copy()
event_df[uni_events] = 0
for ev in uni_events:
#     print(event_df.loc[calendar_raw['event_name_1'] == ev, 'date'])
    event_df.loc[calendar_raw['event_name_1'] == ev, ev] = 1
    event_df.loc[calendar_raw['event_name_2'] == ev, ev] = 1
```

```python
event_df.to_csv('./clean/event_df.csv', index=False)
response_df.to_csv('./clean/response_df.csv', index=False)
```
