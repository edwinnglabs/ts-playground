# Plotting Kernels

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-1, 1, 100)
y = 1 - np.fabs(x) / 1
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 5))
ax.plot(x, y, label='triangular kernel', color='#ffab40')
ax.get_xaxis().set_visible(False)
fig.legend()
fig.tight_layout()
plt.savefig(fname='triangular_kernel.jpg', dpi=300)
```

```python
x = np.linspace(-1, 1, 100)
y = np.exp(-1 * (x) ** 2 / (2 * 0.45 ** 2))
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 5))
ax.plot(x, y, label='gaussian kernel', color='#ffab40')
ax.get_xaxis().set_visible(False)
fig.legend()
fig.tight_layout()
plt.savefig(fname='gaussian_kernel.jpg', dpi=300)
```
